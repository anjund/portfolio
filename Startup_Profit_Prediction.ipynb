```python
# Import libraries and load the dataset
import numpy as np
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv")
df
```

```python
# splitting input and output
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#Show graphically the influence of each factors using seaborn
import seaborn as sns
sns.regplot(x=df['R&D Spend'],y=y,color='red')
```

```python
# One hot encoding
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
x
```

```python
#split data into training data and testing data, testing data= 30%, training data=70%
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
# Now we create a model using LinearRegression and predict the output
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
# Comparing actual values with predicted values
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred})
df1
```

```python
#Slope and intercept
print("intercept",model.intercept_)
print("Slope is",model.coef_)
```

```python
list(zip(x,model.coef_))
```

```python
# Performance measurements
from sklearn.metrics import mean_absolute_error
print("Error is",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print("Percentage error is",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
print("Mean squared error is",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
root_mean=np.sqrt(mean_squared_error(y_test,y_pred))
print("Root mean squared error",root_mean)
```

```python
from sklearn.metrics import r2_score
print("R2 score is",r2_score(y_test,y_pred))
```
